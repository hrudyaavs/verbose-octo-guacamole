```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# 1. Load the dataset into python environment

```python
url = "/content/titanic_dataset (1).csv"
titanic_df = pd.read_csv(url)
```

# 2. Make ‘PassengerId’ as the index column

```python
titanic_df.set_index('PassengerId', inplace=True)
```

# 3. Check the basic details of the dataset

```python
print(titanic_df.info())
print(titanic_df.describe())
```

# 4. Fill in all the missing values present in all the columns in the dataset

```python
titanic_df.fillna(titanic_df.mean(), inplace=True)
```

# 5. Check and handle outliers in at least 3 columns in the dataset

```python
def handle_outliers(column):
    Q1 = titanic_df[column].quantile(0.25)
    Q3 = titanic_df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    titanic_df[column] = titanic_df[column].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

handle_outliers('Age')
handle_outliers('Fare')
handle_outliers('SibSp')
```

### Separate the target variable 'Survived' and exclude non-numeric columns

```python
X = titanic_df.drop(['Survived', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)
y = titanic_df['Survived']
```
